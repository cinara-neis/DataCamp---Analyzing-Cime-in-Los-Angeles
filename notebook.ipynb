```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv(r'C:\Users\Cinara\Desktop\Projetos DataCamp\DataCamp - Analyzing Cime in Los Angeles\dataset\crimes.csv', parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

### Which hour has the highest frequency of crimes?

```python
from dateutil.parser import parse

# Extrair a hora da coluna 'TIME OCC' e contar a frequÃªncia de crimes para cada hora
crimes['Hour'] = pd.to_datetime(crimes['TIME OCC'], format='%H%M', errors='coerce').dt.hour
crime_hour_counts = crimes['Hour'].value_counts()

# Encontrar a hora com a maior frequÃªncia de crimes
peak_crime_hour = crime_hour_counts.idxmax()

# Mostrar a hora com a maior frequÃªncia de crimes
print("Hora com a maior frequÃªncia de crimes:", peak_crime_hour)
```

### Which area has the largest frequency of night crimes

```python
night_crimes_df = crimes[(crimes['Hour'] >= 22) | (crimes['Hour'] < 4)]

# Contar a frequÃªncia de crimes para cada Ã¡rea durante esse perÃ­odo
night_crime_location_counts = night_crimes_df['AREA NAME'].value_counts()

# Identificar a Ã¡rea com a maior frequÃªncia de crimes durante a noite
peak_night_crime_location = night_crime_location_counts.idxmax()

# Armazenar a Ã¡rea com a maior frequÃªncia de crimes durante a noite como uma string
peak_night_crime_location = str(peak_night_crime_location)

# Mostrar a Ã¡rea com a maior frequÃªncia de crimes durante a noite
print("Ãrea com a maior frequÃªncia de crimes durante a noite:", peak_night_crime_location)
```

### Identify the number of crimes committed against victims of different age groups

```python
# Definir os limites de cada faixa etÃ¡ria
age_bins = [0, 18, 26, 35, 45, 55, 65, float('inf')]
age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

# Criar uma nova coluna no DataFrame com a faixa etÃ¡ria de cada vÃ­tima
crimes['Age Group'] = pd.cut(crimes['Vict Age'], bins=age_bins, labels=age_labels, right=False)

# Contar o nÃºmero de crimes para cada faixa etÃ¡ria
victim_ages = crimes['Age Group'].value_counts().sort_index()

# Exibir a sÃ©rie pandas com os resultados
print("NÃºmero de crimes cometidos contra vÃ­timas de diferentes faixas etÃ¡rias:")
print(victim_ages)
```
