```python
# Re-run this cell
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv(r'C:\Users\Cinara\Desktop\Projetos DataCamp\DataCamp - Analyzing Cime in Los Angeles\dataset\crimes.csv', parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

### Which hour has the highest frequency of crimes?

```python
from dateutil.parser import parse

# Extrair a hora da coluna 'TIME OCC' e contar a frequência de crimes para cada hora
crimes['Hour'] = pd.to_datetime(crimes['TIME OCC'], format='%H%M', errors='coerce').dt.hour
crime_hour_counts = crimes['Hour'].value_counts()

# Encontrar a hora com a maior frequência de crimes
peak_crime_hour = crime_hour_counts.idxmax()

# Mostrar a hora com a maior frequência de crimes
print("Hora com a maior frequência de crimes:", peak_crime_hour)
```

### Which area has the largest frequency of night crimes

```python
night_crimes_df = crimes[(crimes['Hour'] >= 22) | (crimes['Hour'] < 4)]

# Contar a frequência de crimes para cada área durante esse período
night_crime_location_counts = night_crimes_df['AREA NAME'].value_counts()

# Identificar a área com a maior frequência de crimes durante a noite
peak_night_crime_location = night_crime_location_counts.idxmax()

# Armazenar a área com a maior frequência de crimes durante a noite como uma string
peak_night_crime_location = str(peak_night_crime_location)

# Mostrar a área com a maior frequência de crimes durante a noite
print("Área com a maior frequência de crimes durante a noite:", peak_night_crime_location)
```

### Identify the number of crimes committed against victims of different age groups

```python
# Definir os limites de cada faixa etária
age_bins = [0, 18, 26, 35, 45, 55, 65, float('inf')]
age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]

# Criar uma nova coluna no DataFrame com a faixa etária de cada vítima
crimes['Age Group'] = pd.cut(crimes['Vict Age'], bins=age_bins, labels=age_labels, right=False)

# Contar o número de crimes para cada faixa etária
victim_ages = crimes['Age Group'].value_counts().sort_index()

# Exibir a série pandas com os resultados
print("Número de crimes cometidos contra vítimas de diferentes faixas etárias:")
print(victim_ages)
```
